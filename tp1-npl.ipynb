```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from collections import Counter
```

```python
stop_words = set(stopwords.words('french'))
stemmer = SnowballStemmer('french')
```

```python
all_words = []
for text in data['text']:
    all_words.extend(word_tokenize(text.lower()))
```

```python
word_freq = Counter(all_words)
```

```python
seuil_mots_rares = 5
rare_words = set([word for word, freq in word_freq.items() if freq < seuil_mots_rares])
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Convertir en minuscules
    stemmed_tokens = [stemmer.stem(token) for token in tokens if token not in stop_words and token not in rare_words]
    return ' '.join(stemmed_tokens)
```

```python
data['text'] = data['text'].apply(preprocess_text)
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
stop_words = set(stopwords.words('french'))
lemmatizer = WordNetLemmatizer()
```

```python
all_words = []
for text in data['text']:
    all_words.extend(word_tokenize(text.lower()))
```

```python
word_freq = Counter(all_words)
```

```python
seuil_mots_rares = 5
rare_words = set([word for word, freq in word_freq.items() if freq < seuil_mots_rares])
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Convertir en minuscules
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words and token not in rare_words]
    return ' '.join(lemmatized_tokens)
```
